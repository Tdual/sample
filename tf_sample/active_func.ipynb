```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.linspace(-10, 10, 400)
y = 1/(1+np.exp(-x))
plt.plot(x, y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
y1 = np.tanh(x)
plt.plot(x, y1)
plt.ylim(-1.1,1.1)
plt.show()
```

```python
def ReLU(x):
    return x * (x > 0)
y2 = ReLU(x)
plt.plot(x, y2)
plt.ylim(-0.3,5)
plt.show()
```
