```python
from numpy.random import randn, rand
import numpy as np 
import scipy.stats as stats
import pandas
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
normal = randn(1000) 
comparison = rand(1000) 
```

```python
plt.hist(normal, bins=100)
plt.show()
```

```python
plt.hist(comparison, bins=100)
plt.show()
```

## Quantile-Quantile Plot
実データ分布のQutantileと理論分布のQutantileを比べることで理論分布と合っているかを確かめる。

### normal

```python
stats.probplot(normal, dist="norm", plot=plt)
plt.show()
```

```python
stats.probplot(comparison, dist="norm", plot=plt)
plt.show()
```

### beta

```python
b = stats.beta(3, 20).rvs(size=500)
plt.hist(b, bins=100)
plt.show()
```

```python
stats.probplot(b, dist=stats.beta, sparams=(3, 20,), plot=plt)
plt.show()
```

```python
stats.probplot(normal, dist=stats.beta, sparams=(3, 20,), plot=plt)
plt.show()
```
