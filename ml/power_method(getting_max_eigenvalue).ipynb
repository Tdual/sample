```python
import numpy as np

def power_method(matrix):
    
    eigenvector = np.ones(len(matrix))
    eigenvalue = 0.0

    if len(eigenvector) == 1:
        return eigenvector


    while True:
        eigenvector_next = np.dot(matrix, eigenvector)
        
        eigenvalue_next =  np.dot(eigenvector_next, eigenvector_next) / np.dot(eigenvector_next, eigenvector)
        print("eigenvalue: ", eigenvalue)
    
        eigenvector_next = eigenvector_next / np.linalg.norm(eigenvector_next)
        if np.allclose(eigenvalue, eigenvalue_next):
            return eigenvalue, eigenvector_next

        eigenvector = eigenvector_next
        eigenvalue =  eigenvalue_next 
```

```python
b = np.array([[1, 1, 2],[0, 2, -1],[0, 0,3]])
b
```

```python
power_method(b)
```
